```python
#leer la tabla de la web con panda
import pandas as pd

linkToFile='https://docs.google.com/spreadsheets/d/e/2PACX-1vSPVJIoDhupNyUpYD65FSHt6kMmF_KaxE2RD6CV1FOMP4vVw7WfKQMqpqo0ieD9vhMAmYP1NpBmw0od/pub?gid=658985149&single=true&output=csv'
dataUN=pd.read_csv(linkToFile)
#abrir la tabla
dataUN
```

```python
#para conocer el tipo de datos de la tabla
#object: texto
#float64: numeros con puntos
dataUN.info()
```

```python
#con la funcion .iloc eliminamos las filas que no contienen informacion de los paises
dataUN_1=dataUN.iloc[7:201,:]
dataUN_1
```

```python
dataUN_1.columns.to_list()
```

```python
#con la funcion . loc selecciono las columnas con las que me quiero quedar (las que tienen informacion completa)
keep=['Back','Table 1. Human Development Index and its components','Unnamed: 2','Unnamed: 4','Unnamed: 6','Unnamed: 8','Unnamed: 10','Unnamed: 12','Unnamed: 14']
dataUN_1=dataUN_1.loc[:, keep]
```

```python
#renombro los encabezados de la tabla
renombrar={'Back':'Nº','Table 1. Human Development Index and its components':'Country','Unnamed: 2':'Human Development Index (HDI)','Unnamed: 4':'Life expectancy at birth','Unnamed: 6':'Expected years of schooling','Unnamed: 8':'Mean years of schooling','Unnamed: 10':'Gross national income (GNI) per capita','Unnamed: 12':'GNI per capita rank minus HDI rank','Unnamed: 14':'HDI rank for 2020'}
dataUN_1.rename(columns=renombrar,inplace=True)
dataUN_1
```

```python
#detectar la presencia de simbolos
set(dataUN_1['Gross national income (GNI) per capita'].str.extractall('(\W)').loc[:,0].to_list()) - set('.')
```

```python
#eliminar las comas
dataUN_1['Gross national income (GNI) per capita'] = dataUN_1['Gross national income (GNI) per capita'].str.replace(',', '')
dataUN_1
```

```python
import os

try:
    os.mkdir('data')
except FileExistsError:
    pass

dataUN_1.to_csv(os.path.join("data", "dataUN_1.csv"), index=False)
```
